# 線形ガウス状態空間モデルとカルマンフィルタ
## 概要

- [モデル]:線形ガウス状態空間モデル
- [状態の推定方法]:カルマンフィルタ(/or 散漫カルマンフィルタ) =>  　平滑化
- [パラメタの推定]: 最尤推定
- [用いるソフト]:R
- [使用するパッケージ]:ｄｌｍ, KFAS

## 制約
- 　 *線形* かつ *正規分布*に限る

## 表現：状態方程式/観測方程式
$$
x_{t} = T_{t}x_{t-1} + R_{t}ε_{t}
$$
$$
y_{t } = Z_{t}x_{t} +　ε_{t}
$$

## 状態推定
- フィルタリングとは　=　手に入った観測値をもちいて, 予測された状態値を補正する

1. 状態方程式をもとに次の状態値を予測する
2. 予測した状態値を補正する
3. 状態の補正がおわった後に再度状態を予測する

## 状態推定:平滑化
- 全てのデータが揃ってから状態の補正を行う

## パラメタ推定
- 観測データを元に状態値を補正するために=>過程誤差の分散, 観測誤差の分散の大きさの情報が必要
- 誤差の分散が大きいなら情報として信用できない
- => 過程誤差と観測誤差の比率を勘案しつつフィルタリング（状態の補正）を行う
- 過程誤差と観測誤差を最尤推定を用いて推測する

## ローカルレベルモデル
- 観測誤差と過程誤算の両方を認めたモデル
- ホワイトノイズの累積和(ランダムウォーク)に観測誤差が加わったもの
- 予測にそこまで関わらない
$$
μ_{t} = μ_{t-1} + ε_{t}
$$
$$
y_{t} = μ_{t} + v_{t}
$$

## ローカル線形トレンドモデル
- ローカルレベルモデルにトレンド成分を加えたもの
- トレンド成分も変化することをかみしている

$$
δ_{t} = δ_{t-1} + ε_{t}
$$
$$
μ_{t} =  μ_{t-1} + δ_{t} + w_{t}
$$
$$
y_{t} = 　μ_{t}　+ v_{t}
$$ 

##  基本構造系時系列モデル
- 時系列データ=トレンド + 周期的変動 + ホワイトノイズ
- 状態空間モデルで表すことが可能


# 状態推定:カルマンフィルタ
## カルマンフィルタとカルマンゲイン
- 補正後の状態 = 補正前の状態 + カルマンゲイン + （実際の観測値 - 予想された観測値）
- 補正後の状態をフィルタ化推定量　と呼ぶ
-　カルマンゲイン = 状態の予測誤差の分散/ (状態の予測誤差の分散 + 観測誤差の分散)

## カルマンゲイン
- 常に１以下で設定されている
- 状態= 前時点の状態+ 過程誤差
- 観測値 = 状態 + 観測誤差

### 計算手順
1. 状態を予想する. ローカルレベルなので前期と変化しない
2. フィルタリング：(状態の補正)

## 数式で見るカルマンフィルタ
- 数式を用いてローカルレベルモデルのカルマンフィルタの計算の流れを再度検証する
$$
μ_{t} = μ_{t-1} + w_{t}
$$
$$
y_{t} = μ_{t} + v_{t}
$$

# 状態推定:散漫カルマンフィルタ
## 状態の初期値問題
状態の予測 =　前の状態

- 状態の予測誤差の分散 = 前期の状態の予測誤差の分散 + 過程誤差の分散
- 状態の予測には前回の値が必要
